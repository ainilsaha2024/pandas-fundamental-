# Reading dataset from Github using requests 

```python
import requests
from io import StringIO 
```

```python
url ="https://github.com/rashakil-ds/5-Minutes-to-Pandas/blob/main/Datasets/bike.csv"
response = requests.get(url)
```

```python
csv_content =StringIO(response.text)
```

```python
df1 = pd.read_csv(csv_content)
print(df1)
```

```python
df1.head()
```

```python
df1.info()
```

```python
df1.shape
```

# Reading dataset from Github using web Get 

```python
import wget
```

```python
url = "https://raw.githubusercontent.com/rashakil-ds/5-Minutes-to-Pandas/refs/heads/main/Datasets/bike.csv"

file_path =wget.download(url)
```
